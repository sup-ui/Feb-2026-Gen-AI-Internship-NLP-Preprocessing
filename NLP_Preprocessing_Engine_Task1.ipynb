# **Task 1: Conceptual Understanding**

**1. Difference between "Love" and "love" in NLP**

In Natural Language Processing, "Love" and "love" are treated as different tokens if case normalization is not applied. This is because NLP models often treat text literally.

For example:

•	Love makes people happy

Tokenized without lowercasing:
["Love", "makes", "people", "happy"]

Another sentence:

•	I love music
Tokens:
["I", "love", "music"]

Here Love ≠ love, meaning the vocabulary size increases unnecessarily.
So to avoid this, most NLP pipelines apply lowercasing so both become love.


**2. What happens if stopwords are not removed?**

Stopwords are commonly occurring words such as:

**[the, is, at, in, on, and]**

If stopwords remain in the dataset:

•	The dataset becomes larger and noisier

•	Models spend time learning unimportant words

•	Important keywords get less weight

But in some tasks removing stopwords may remove useful information.


**3. Two real-world scenarios where removing stopwords can be harmful**

•	**Sentiment Detection**

 “The movie was not good”

Removing stopwords might produce:

movie good

The word "not" carries sentiment reversal, so removing it changes meaning.

•	**Search Engine Queries**

“how to apply for passport”

If stopwords are removed:

apply passport

Important context words "how" and "for" are lost, which may reduce search accuracy.



**4. Difference between Stemming and Lemmatization**

•	Stemming is a simple method that removes suffixes from words using predefined rules. It does not check whether the resulting word is grammatically correct or exists in a dictionary. Because of this, the output can sometimes be an incomplete or incorrect word.
 For example, the word "studies" may be reduced to "studi" after stemming because the algorithm simply removes the ending "es".

•	Lemmatization, on the other hand, is a more advanced process that uses vocabulary dictionaries and linguistic analysis to determine the correct base form of a word. It considers the grammatical meaning of the word before transforming it.
For example, the word "studies" would be converted to "study", which is the proper dictionary form.

•	Stemming is faster and computationally simpler, while lemmatization is more accurate but slightly slower because it involves dictionary lookups and language rules.



# **Task 2:Build Advanced Preprocessing Function**

This task focuses on building a preprocessing function that cleans raw text by removing noise such as numbers, URLs, extra spaces and repeated characters.
The function also converts text to lowercase and filters unnecessary tokens to produce clean tokens suitable for NLP models.



```python
# Import required library
import re
def preprocess_text(text):
    # Handle empty input safely
    if text is None or text.strip() == "":
        return [], ""

    # Convert text to lowercase
    text = text.lower()

    # Remove URLs (http, https, www)
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (example: sooooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove punctuation and special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize text
    tokens = text.split()

    # Remove very short tokens except meaningful ones
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Reconstruct cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence

#test cases
test_sentences = [
    "I have 2 dogs",
    "This   is good",
    "Soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
    "Email me at test123@mail.com",
    "Noooo this is baaad",
    "OK OK I understand",
    "My score is 95 in exam",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 50)
```

# **Task 3: Stress Testing**
This task tests the preprocessing function on different types of noisy text such as emojis, numbers, slang, repeated characters, URLs and mixed-case sentences to ensure the function works correctly in diverse real-world scenarios.


```python
# Import required library
import re

# Preprocessing function from Task 2
def preprocess_text(text):

    # Handle empty input
    if text is None or text.strip() == "":
        return [], ""

    # Convert text to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (example: sooooo -> so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters and emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize text
    tokens = text.split()

    # Remove very short tokens except "no" and "not"
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    # Reconstruct cleaned sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence


# test cases

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


# preprocessing and ouput

for sentence in test_sentences:

    tokens, cleaned_sentence = preprocess_text(sentence)

    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 50)
```

# **Task 4: Token Analytics**
This task analyzes each cleaned sentence by calculating the total number of tokens, the number of unique tokens and the average token length to understand the quality of the preprocessing results.

```python
import numpy as np
# Loop through each sentence and calculate token statistics
for i, sentence in enumerate(test_sentences):
    tokens, cleaned_sentence = preprocess_text(sentence)
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    # Calculate average token length safely
    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    print("Original Sentence:", sentence)
    print("Cleaned Sentence:", cleaned_sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length,2))
    print("-"*50)
```

# **Task 5: Frequency Analysis**
This task combines all tokens from the processed sentences and uses the Counter class to identify the most frequent and least frequent words.

```python
from collections import Counter
# List to store all tokens
all_tokens = []
# Collect tokens from each sentence
for sentence in test_sentences:
    tokens, cleaned_sentence = preprocess_text(sentence)
    # Add tokens to the list
    all_tokens.extend(tokens)

# Create a Counter object
word_freq = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))
print()
# Top 5 least frequent words
print("Top 5 Least Frequent Words:")
print(word_freq.most_common()[-5:])
```

# **Task 6: Build Full Pipeline**
This task creates a complete preprocessing pipeline that processes a list of sentences and returns all tokens and cleaned sentences in a structured dictionary.

```python
# Full preprocessing pipeline
def full_pipeline(text_list):

    # List to store all tokens
    all_tokens = []

    # List to store cleaned sentences
    clean_sentences = []

    # Process each sentence in the input list
    for text in text_list:

        # Apply preprocessing
        tokens, cleaned_sentence = preprocess_text(text)

        # Add tokens to the token list
        all_tokens.extend(tokens)

        # Add cleaned sentence to the sentence list
        clean_sentences.append(cleaned_sentence)

    # Return results in dictionary format
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Example usage
result = full_pipeline(test_sentences)
print(result)
```
